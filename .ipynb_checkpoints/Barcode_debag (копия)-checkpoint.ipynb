```python
image_test = cv2.imread('photo/1.png') # open image and view
plt.imshow(image_test)
plt.show()
#cv2.imwrite('photo/out.png',image_test)
```

```python
# MSER algoritm 
# работает только с чётким
image_mser = image_test.copy()
mser = cv2.MSER_create(5, 60, 14400, 0.25, 0.2, 200, 1.01, 0.003, 5) # inicialisation MSER and his parameters

msers, bboxes = mser.detectRegions(image_mser)
out_image = np.zeros(image_mser.shape,dtype = np.uint8)
out_image = np.full(image_mser.shape,(255,255,255),dtype = np.uint8)
for i in msers:
    out_image = cv2.polylines(out_image, [i], 1, (0, 0, 0))
plt.imshow(out_image) # view image
plt.show()
```

```python
b =bboxes.copy()
b[:,2:] +=b[:,:2]
v = np.hstack([np.min(b[:,:2],axis=0),np.amax(b[:,2:],axis = 0)])
plt.imshow(cv2.rectangle(out_image,tuple(v[:2]),tuple(v[2:]),(0,0,255),2))
plt.show()
```

```python
kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 10))
closed2 = cv2.morphologyEx(image_thresh, cv2.MORPH_CLOSE, kernel1)


plt.imshow(closed2)
plt.show()

```

```python
kernel = np.ones((1,20),np.uint8)
blackhat = cv.morphologyEx(closed2, cv.MORPH_BLACKHAT, kernel)
closed = cv2.erode(blackhat, None, iterations = 1)
ret,thresh2 = cv.threshold(closed,0,255,cv.THRESH_BINARY_INV)

#b = thresh2.copy()
#b[:,2:] +=b[:,:2]
#v = np.hstack([np.min(b[:,:2],axis=0),np.amax(b[:,2:],axis = 0)])

plt.imshow(blackhat)
plt.show()
plt.imshow(closed)
plt.show()
plt.imshow(thresh2)
plt.show()

#plt.imshow(cv2.rectangle(closed,tuple(v[:2]),tuple(v[2:]),(0,0,255),2))
#plt.show()
```

```python

hough_image = thresh2.copy()    #Hough transformation
gray = cv2.cvtColor(hough_image,0)
edges = cv2.Canny(gray,50,150,apertureSize = 3)
lines = cv2.HoughLines(edges,1,np.pi/180,42)

__lines = lines[:,0,:].copy()
for rho,theta in __lines:
    
    a = np.cos(theta)
    b = np.sin(theta)
    c = 1000
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + c*(-b))
    y1 = int(y0 + c*(a))
    x2 = int(x0 - c*(-b))
    y2 = int(y0 - c*(a))


    
    plt.imshow(cv2.line(hough_image,(x1,y1),(x2,y2),(0,0,255),2))
plt.show()

    


#cv2.imwrite('photo/out.png',hough_image) #command of write image
```

# Преобразование хафа

```python
img = cv2.imread('photo/15.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,150,apertureSize = 3)
minLineLength = 100
maxLineGap = 10
lines = cv2.HoughLinesP(edges,1,np.pi/180,100,minLineLength,maxLineGap)
for x1,y1,x2,y2 in lines[0]:
    cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)

plt.imshow(img)

plt.show()
```
