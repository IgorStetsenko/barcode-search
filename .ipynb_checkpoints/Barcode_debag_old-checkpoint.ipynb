```python
photo_test = 'photo/4.png'
gray_color = 200
image_test = cv2.imread(photo_test)
image_test[image_test >= gray_color]= 255
image_test[image_test < gray_color] = 0

plt.imshow(image_test)

plt.show()
#cv2.imwrite('out.png', image_test)
```

```python

mser = cv2.MSER_create(5, 1, 14400, 0.25, 0.2, 315, 1.01, 0.003, 1) # inicialisation MSER and his parameters
plt.imshow(image_test) # view image
plt.show()
```

```python
msers, bboxes = mser.detectRegions(image_test)
plt.imshow(image_test) # view image
plt.show()
cv2.imwrite('out_im.png',image_test)
```

```python
image_test = cv2.imread('out_im.png')
out_image = np.zeros(image_test.shape,dtype = np.uint8)
out_image = np.full(image_test.shape,(255,255,255),dtype = np.uint8)
for i in msers:
    out_image = cv2.polylines(out_image, [i], 1, (0, 0, 0))
plt.imshow(out_image)

plt.show()
```

```python

img = out_image
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Set up the detector and configure its params.
params = cv2.SimpleBlobDetector_Params()
params.minDistBetweenBlobs = 0
params.filterByColor = False
params.blobColor = 255
params.filterByArea = False
params.minArea = 10
params.maxArea = 3000
params.filterByCircularity = False
params.filterByConvexity = False
params.filterByInertia = True
params.minInertiaRatio = 0.01
params.maxInertiaRatio = 1
detector = cv2.SimpleBlobDetector_create(params)

# Detect blobs.
keypointsb = detector.detect(img)

# Draw detected blobs as red circles.
im_with_keypoints = cv2.drawKeypoints(img, keypointsb, np.array([]), (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Show keypoints
#cv2.imwrite('photo/1.png',im_with_keypoints)
```

```python
img = cv2.imread('photo/1.png')
mser = cv2.MSER_create()
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #Converting to GrayScale
gray_img = img.copy()

regions = mser.detectRegions(gray)
hulls = [cv2.convexHull(p.reshape(-1, 1, 2)) for p in regions]
cv2.polylines(gray_img, hulls, 1, (0, 0, 255), 2)
cv2.imwrite('photo/111.png', gray_img) #Saving
```
