```python
# MSER algoritm 
# работает только с чётким
image_mser = image_test.copy()
mser = cv2.MSER_create(5, 60, 14400, 0.25, 0.2, 415, 1.01, 0.003, 5) # inicialisation MSER and his parameters

msers, bboxes = mser.detectRegions(image_mser)
out_image = np.zeros(image_mser.shape,dtype = np.uint8)
out_image = np.full(image_mser.shape,(255,255,255),dtype = np.uint8)
for i in msers:
    out_image = cv2.polylines(out_image, [i], 1, (0, 0, 0))
plt.imshow(out_image) # view image
plt.show()
```

```python
b =bboxes.copy()
b[:,2:] +=b[:,:2]
v = np.hstack([np.min(b[:,:2],axis=0),np.amax(b[:,2:],axis = 0)])
plt.imshow(cv2.rectangle(out_image,tuple(v[:2]),tuple(v[2:]),(0,0,255),2))
plt.show()
```

```python
image_bin = blur.copy() #Binarization Otsu
ret,image_thresh = cv2.threshold(image_bin,150,255,cv2.THRESH_BINARY_INV) #binarization and inversion
#cv2.imshow("Barcode image", image_thresh)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.imshow(image_thresh)
plt.show()
```

# Формирование массива периметров

```python
perimeter = np.array([])# create array
for item in contours:
    perimeter = np.append(perimeter, cv2.arcLength(item,True))
    
    
    
```

```python
perimeter #массив периметров
```

```python
perimeter.mean() #среднее по всем элементам  (не очень)
```

```python
plt.hist(perimeter,255) #смотрим распределение периметров на гистограмме
plt.show()
```

```python
image_unique = np.unique(perimeter,return_counts=True) #уникальные элементы #второй массив - количество вхождений
image_unique
```

```python
image_unique_element_index = [image_unique[1]==image_unique[1].max()] #количество уникальных элементов, 
                                                                      #соответствующее максимальному вхождению элемента
```

```python
image_unique_element_index
```

```python
image_unique_mean = image_unique[0].mean()  #среднее по разбросу. Взять и прибавить 40% 
```

```python
image_unique_mean 
```

```python
image_unique_diap = np.logical_and(image_unique[0]<image_unique_mean*1.7, image_unique[0]>image_unique_mean*0.3)
#устанавливаем разброс +- 70% от среднего
```

```python
image_unique_diap
```

```python
image_unique[0][image_unique_diap] #входящие диапазон элементы
```

```python
index = np.array([])
```

```python
#создаём массив индексов контуров
for i in image_unique[0][image_unique_diap]:   #индекс по элементу
    perimeter_index = np.take(contours,np.where(perimeter == i))
    index = np.append(index, perimeter_index)
```

```python
#index
```

```python
for contours_coordinate in index:# print conturs
    x1 = contours_coordinate.min(axis = 0)
    y1 = contours_coordinate.max(axis = 0)
    x2 = contours_coordinate.max(axis = 0)
    y2 = contours_coordinate.min(axis = 0)
    zones = cv2.rectangle(image_test,(x1[0,0],y1[0,1]),(x2[0,0],y2[0,1]),(0,255,255),2)
    plt.imshow(zones)

plt.show()
```
